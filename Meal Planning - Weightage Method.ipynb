```python
# Create new cols -  meal info dictionaries and protein lists
def get_meal_info(segment, cabin_class, date, meal_time):
    # Filter meal_df for matching rows
    matching_meals = meal_df[
        (meal_df['segment'] == segment) & 
        (meal_df['cabin_class'] == cabin_class) & 
        (meal_df['segment_local_departure_date'] == date) & 
        (meal_df['meal_time'] == meal_time)
    ]
    
    if matching_meals.empty:
        return [], []
    
    # Create list of dictionaries for meal info
    meal_info_list = []
    protein_list = []
    
    for _, meal_row in matching_meals.iterrows():
        meal_dict = {
            "meal_name": meal_row['meal_name'],
            "protein": meal_row['meal_pref']
        }
        meal_info_list.append(meal_dict)
        protein_list.append(meal_row['meal_pref'])
    
    # Remove duplicates and sort alphabetically for deterministic ordering
    # This ensures consistent tie-breaking in largest remainder method
    unique_proteins = sorted(list(set(protein_list)))
    
    return meal_info_list, unique_proteins
```

```python
# get meals available and proteins availabe for every feat group...
daily_feature_counts[['meals_available', 'proteins_available']] = daily_feature_counts.apply(
    lambda row: pd.Series(get_meal_info(
        row['segment'], 
        row['cabin_class'], 
        row['segment_local_departure_date'], 
        row['meal_time']
    )), axis=1
)
```

## Code functions : restrict the proteins and normalize the probability

Some proteins are not available in the flight. So restrict to the proteins available in flight.

These code functions will be used later

nationality_w , age_w, destination_w , mealtime_w 

```python
# This function will look through each df row's selected feature (eg. nationality / age / etc)
#and... restricts / takes only the available protein types and then normalize the probability to 1 


def get_restricted_probabilities(feature_value, weight_df, feature_column, available_proteins):
    
    # Get the row for this feature value...in the weight_df 
    feature_row = weight_df[weight_df[feature_column] == feature_value]
    
    if feature_row.empty:
        # Return equal probabilities if feature not found
        equal_prob = 1.0 / len(available_proteins)
        return {protein: equal_prob for protein in available_proteins}
    
    # Get probabilities for available proteins only
    protein_columns = ['Pork', 'Chicken', 'Beef', 'Seafood', 'Lamb', 'Vegetarian']
    available_probs = {}
    total_prob = 0
    
    for protein in available_proteins:
        if protein in protein_columns:
            prob = feature_row[protein].iloc[0]
            available_probs[protein] = prob
            total_prob += prob
    
    # normalize probabilities to get sum = 1
    if total_prob > 0:
        normalized_probs = {protein: prob/total_prob for protein, prob in available_probs.items()}
    else:
        # IF all probabilities are 0, assign equal probabilities
        equal_prob = 1.0 / len(available_proteins)
        normalized_probs = {protein: equal_prob for protein in available_proteins}
    
    return normalized_probs

```

```python
def get_restricted_probabilities_with_weekday(nationality_code, weekday, nationality_weight_df, available_proteins):

    # Filter for matching nationality and weekday
    feature_row = nationality_weight_df[
        (nationality_weight_df['nationality_code'] == nationality_code) & 
        (nationality_weight_df['day_of_week'] == weekday)
    ]
    
    if feature_row.empty:
        # Return equal probabilities if not found
        equal_prob = 1.0 / len(available_proteins)
        return {protein: equal_prob for protein in available_proteins}
    
    # Get probabilities for available proteins only
    protein_columns = ['Pork', 'Chicken', 'Beef', 'Seafood', 'Lamb', 'Vegetarian']
    available_probs = {}
    total_prob = 0
    
    for protein in available_proteins:
        if protein in protein_columns:
            prob = feature_row[protein].iloc[0]
            available_probs[protein] = prob
            total_prob += prob
    
    # Normalize probabilities to sum to 1
    if total_prob > 0:
        normalized_probs = {protein: prob/total_prob for protein, prob in available_probs.items()}
    else:
        # If all probabilities are 0, assign equal probabilities
        equal_prob = 1.0 / len(available_proteins)
        normalized_probs = {protein: equal_prob for protein in available_proteins}
    
    return normalized_probs
```

## Start calculation 

(Probaility from llm) x (Customer feature set)

```python
# Parameter weights
W1 = 0.40  # Nationality
W2 = 0.20  # Age Group
W3 = 0.25  # Destination
W4 = 0.15  # Meal Time
```

```python
# Calculate final probabilities for each row
final_results = []

for idx, row in daily_feature_counts.iterrows():
    available_proteins = row['proteins_available']
    
    if not available_proteins or len(available_proteins) == 0:
        continue  
    
    #  normalize probabilities for eac feature
    nat_probs = get_restricted_probabilities_with_weekday(row['nationality_code'], row['weekday'], nationality_w, available_proteins)
    age_probs = get_restricted_probabilities(row['age_group'], age_w, 'age_group', available_proteins)
    dest_probs = get_restricted_probabilities(row['destination_region'], destination_regions_unique, 'destination_region', available_proteins)
    meal_probs = get_restricted_probabilities(row['meal_time'], mealtime_w, 'meal_time', available_proteins)
    
    # weighted probabilities for EACH protein
    weighted_protein_probs = {}
    
    for protein in available_proteins:
        weighted_prob = (nat_probs.get(protein, 0) * W1 + 
                        age_probs.get(protein, 0) * W2 + 
                        dest_probs.get(protein, 0) * W3 + 
                        meal_probs.get(protein, 0) * W4)
        weighted_protein_probs[protein] = weighted_prob
    
    # normalize probabilities to sum to 1 --FINAL
    total_final_prob = sum(weighted_protein_probs.values())
    if total_final_prob > 0:
        final_protein_probs = {protein: prob/total_final_prob for protein, prob in weighted_protein_probs.items()}


    ###############CALCULATE PASSENGERS##########################################
    # Calculate passenger counts using largest remainder method - NO NEGATIVES POSSIBLE
    protein_counts = {}
    remainders = {}
    
    # step 1 - Calculate exact counts and floor them (never negative)
    for protein in available_proteins:
        count_exact = row['passenger_count'] * final_protein_probs[protein]
        floor_count = int(count_exact)  # floor (always more than 0)
        remainder = count_exact - floor_count
        
        protein_counts[protein] = {
            'probability': round(final_protein_probs[protein], 4),
            'count_exact': round(count_exact, 2),
            'count_final': floor_count  # start with floor (more than 0)
        }
        remainders[protein] = remainder
    
    # step 2 -- calculate remaining passengers to distribute
    total_allocated = sum([protein_counts[p]['count_final'] for p in available_proteins])
    remaining_passengers = row['passenger_count'] - total_allocated
    
    # Step 3-  distribute remaining passengers to proteins with largest remainders
    if remaining_passengers > 0:
        # sort proteins by remainder (largest first)
        sorted_proteins = sorted(available_proteins, 
                               key=lambda p: remainders[p], reverse=True)
        
        # give 1 passenger each to proteins with largest remainders
        for i in range(min(remaining_passengers, len(sorted_proteins))):
            protein = sorted_proteins[i]
            protein_counts[protein]['count_final'] += 1
    
    # Step 4 - Final check (should always be integer and non negative)
    total_allocated = sum([protein_counts[p]['count_final'] for p in available_proteins])
    passenger_difference = total_allocated - row['passenger_count']
    tally_matches = (passenger_difference == 0)
    
    # all protein counts are confirm more than 0 because:
    # -we start with floor (so like more than 0)
    # - we do not substract, only add passengers
    
    
    #  result dictionary
    result_dict = {
        'segment': row['segment'],
        'cabin_class': row['cabin_class'],
        'segment_local_departure_date': row['segment_local_departure_date'],
        'weekday': row['weekday'],
        'feature_set': row['feature_set'],
        'nationality_code': row['nationality_code'],
        'age_group': row['age_group'],
        'destination_region': row['destination_region'],
        'meal_time': row['meal_time'],
        'passenger_count': row['passenger_count'],
        'available_proteins': available_proteins,
        'meals_available': row['meals_available'],
        #  CHECK ----> NEW COLUMNS
        'total_allocated': total_allocated,
        'passenger_difference': passenger_difference,
        'tally_matches': tally_matches
        
    }
    
    # probabilities (normalized)
    for protein in available_proteins:
        result_dict['weighted_protein_probs'] = weighted_protein_probs
        result_dict[f'nat_{protein.lower()}_norm_prob'] = round(nat_probs.get(protein, 0), 4)
        result_dict[f'age_{protein.lower()}_norm_prob'] = round(age_probs.get(protein, 0), 4)
        result_dict[f'dest_{protein.lower()}_norm_prob'] = round(dest_probs.get(protein, 0), 4)
        result_dict[f'meal_{protein.lower()}_norm_prob'] = round(meal_probs.get(protein, 0), 4)
        result_dict[f'{protein.lower()}_final_probability'] = protein_counts[protein]['probability']
        result_dict[f'{protein.lower()}_count_exact'] = protein_counts[protein]['count_exact']
        result_dict[f'{protein.lower()}_count_final'] = protein_counts[protein]['count_final']
    
    final_results.append(result_dict)
```

## Transpose Results Table

```python
# Initialize transposed table
transposed_results = []

for idx, row in aggregated_results.iterrows():
    meals_available = row['meals_available']
    total_passengers = row['passenger_count']
    
    # get all protein count columns that have values...a lotof is NAN value because that meal wasnt present
    protein_counts = {}
    for col in aggregated_results.columns:
        if col.endswith('_count_final'):
            protein_type = col.replace('_count_final', '').title()
            count = row[col]
            if count > 0:  # Only include proteins with counts > 0
                protein_counts[protein_type] = count
    
    # create row for EACH meal in meals_available
    for meal_info in meals_available:
        meal_name = meal_info['meal_name']
        meal_protein = meal_info['protein'].title()
        
        # get the count for this particular protein type (0 if not found)
        count = protein_counts.get(meal_protein, 0)
        
        # calc percentage
        percentage = round((count / total_passengers) * 100, 2) if total_passengers > 0 and count > 0 else 0
        
        # check if this protein type appears multiple times in meals_available
        protein_meal_count = sum(1 for meal in meals_available if meal['protein'].title() == meal_protein)
        protein_repeated = protein_meal_count > 1
        
        transposed_results.append({
            'segment': row['segment'],
            'cabin_class': row['cabin_class'],
            'segment_local_departure_date': row['segment_local_departure_date'],
            'meal_time': row['meal_time'],
            'passenger_count': row['passenger_count'],
            'meal_name': meal_name,
            'protein_type': meal_protein,
            'count': count,
            'percentage': percentage,
            'protein_repeated': protein_repeated,
            'meals_available': meals_available
        })

```

```python
transposed_df = pd.DataFrame(transposed_results)

```
